```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv('Wholesale customers data.csv')
columns_to_normalize = [col for col in df.columns if col not in ['Channel', 'Region']]
```

```python
normalize_scaler = MinMaxScaler()
second_normalize_scaler = MinMaxScaler(feature_range=(-1,1))
standard_scaler = StandardScaler()
```

```python
df[columns_to_normalize] = standard_scaler.fit_transform(df[columns_to_normalize])
df.to_csv('standarized.csv', index=False)
```
