Dla P1:
 	3.1. Zastosowanie algorytmu k-średnich dla liczby grup równej wartości początkowej (użytej przy inicjowaniu zbiorów - 3). 
	3.2. Zastosowanie algorytmu hierachicznego dla liczby grup równej wartości początkowej (użytej przy inicjowaniu zbiorów), metoda łączenia: Ward, metryka: euclidean. 
	3.3. Zastosowanie algorytmu DBSCAN, promień sąsiedztwa wynosi 0.5, minimalna liczba punktów w odległości epsilon to 8.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import matplotlib.pyplot as plt

# Dla p1:
# 3.1. Zastosowanie algorytmu k-średnich dla liczby grup równej wartości początkowej (użytej przy inicjowaniu zbiorów - 3).
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans_labels = kmeans.fit_predict(set_1[0])

# 3.2. Zastosowanie algorytmu hierarchicznego dla liczby grup równej wartości początkowej (użytej przy inicjowaniu zbiorów), metoda łączenia: Ward, metryka: euclidean.
agg_clustering = AgglomerativeClustering(n_clusters=3, linkage="ward", affinity="euclidean")
agg_labels = agg_clustering.fit_predict(set_1[0])

# 3.3. Zastosowanie algorytmu DBSCAN, promień sąsiedztwa wynosi 0.5, minimalna liczba punktów w odległości epsilon to 8
dbscan = DBSCAN(eps=0.5, min_samples=8)
dbscan_labels = dbscan.fit_predict(set_1[0])

# Wykresy wyników
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.scatter(set_1[0][:, 0], set_1[0][:, 1], c=kmeans_labels, cmap='viridis')
plt.title("K-Means Clustering")

plt.subplot(132)
plt.scatter(set_1[0][:, 0], set_1[0][:, 1], c=agg_labels, cmap='viridis')
plt.title("Agglomerative Clustering")

plt.subplot(133)
plt.scatter(set_1[0][:, 0], set_1[0][:, 1], c=dbscan_labels, cmap='viridis')
plt.title("DBSCAN Clustering")

plt.show()
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Dla p2:

# 4.1. Zastosowanie algorytmu k-średnich dla liczby grup równej 8.
kmeans = KMeans(n_clusters=8, random_state=0)
kmeans_labels = kmeans.fit_predict(set_2[0])

# 4.2. Zastosowanie metody inercji (n do 16) do wyznaczenia optymalnej liczby skupień.
inertia_values = []
for n_clusters in range(2, 17):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(set_2[0])
    inertia_values.append(kmeans.inertia_)

# 4.3. Zastosowanie dla algorytmu k-średnich Silhouette score (n do 16) do wyznaczenia optymalnej liczby skupień.
silhouette_scores = []
for n_clusters in range(2, 17):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans_labels = kmeans.fit_predict(set_2[0])
    silhouette_scores.append(silhouette_score(set_2[0], kmeans_labels))

# 4.4. Zastosowanie algorytmu hierarchicznego dla liczby grup równej wartości początkowej (użytej przy inicjowaniu zbiorów), metoda łączenia: Ward, metryka: euclidean.
agg_clustering = AgglomerativeClustering(n_clusters=8, linkage="ward", affinity="euclidean")
agg_labels = agg_clustering.fit_predict(set_2[0])

# 4.5. Zastosowanie dla algorytmu hierarchicznego Silhouette score (n do 16) do wyznaczenia optymalnej liczby skupień.
silhouette_scores_agg = []
for n_clusters in range(2, 17):
    agg_clustering = AgglomerativeClustering(n_clusters=n_clusters, linkage="ward", affinity="euclidean")
    agg_labels = agg_clustering.fit_predict(set_2[0])
    silhouette_scores_agg.append(silhouette_score(set_2[0], agg_labels))

# 4.6. Zastosowanie algorytmu DBSCAN, promień sąsiedztwa wynosi 2, minimalna liczba punktów w odległości epsilon to 20.
dbscan = DBSCAN(eps=2, min_samples=20)
dbscan_labels = dbscan.fit_predict(set_2[0])

# Wykresy wyników
plt.figure(figsize=(16, 12))

plt.subplot(231)
plt.scatter(set_2[0][:, 0], set_2[0][:, 1], c=kmeans_labels, cmap='viridis')
plt.title("K-Means Clustering (8 clusters)")

plt.subplot(232)
plt.plot(range(2, 17), inertia_values, marker='o', linestyle='-')
plt.title("Elbow Method for K-Means")

plt.subplot(233)
plt.plot(range(2, 17), silhouette_scores, marker='o', linestyle='-')
plt.title("Silhouette Scores for K-Means")

plt.subplot(234)
plt.scatter(set_2[0][:, 0], set_2[0][:, 1], c=agg_labels, cmap='viridis')
plt.title("Agglomerative Clustering (8 clusters)")

plt.subplot(235)
plt.plot(range(2, 17), silhouette_scores_agg, marker='o', linestyle='-')
plt.title("Silhouette Scores for Agglomerative Clustering")

plt.subplot(236)
plt.scatter(set_2[0][:, 0], set_2[0][:, 1], c=dbscan_labels, cmap='viridis')
plt.title("DBSCAN Clustering")

plt.show()
```
