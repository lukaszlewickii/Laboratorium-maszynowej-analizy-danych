```python
from sklearn.datasets import make_moons, make_circles, make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. Przygotowanie zbior√≥w danych:

#1.1.
set_1 = make_blobs(n_samples=1000, centers=3, cluster_std=0.99, random_state=3042019)
#1.2.
set_2 = make_blobs(n_samples=2000, centers=8, cluster_std=2.5, random_state=42)
#1.3.
set_3 = make_moons(n_samples=1000, noise=5, random_state=0)
#1.4.
set_4 = make_circles(n_samples=1000, random_state=0)
```

```python
#2. Wizualizacja ka≈ºdego zbioru danych.

# Ustawienia dla subplott√≥w
plt.figure(figsize=(16, 12))

# Zestaw danych 1.1.
set_1 = make_blobs(n_samples=1000, centers=3, cluster_std=0.99, random_state=3042019)
plt.subplot(221)
plt.scatter(set_1[0][:, 0], set_1[0][:, 1], c=set_1[1], cmap='viridis')
plt.title('Zestaw danych 1.1.')

# Zestaw danych 1.2.
set_2 = make_blobs(n_samples=2000, centers=8, cluster_std=2.5, random_state=42)
plt.subplot(222)
plt.scatter(set_2[0][:, 0], set_2[0][:, 1], c=set_2[1], cmap='viridis')
plt.title('Zestaw danych 1.2.')

# Zestaw danych 1.3.
set_3 = make_moons(n_samples=1000, noise=0.05, random_state=0)
plt.subplot(223)
plt.scatter(set_3[0][:, 0], set_3[0][:, 1], c=set_3[1], cmap='viridis')
plt.title('Zestaw danych 1.3.')

# Zestaw danych 1.4.
set_4 = make_circles(n_samples=1000, random_state=0)
plt.subplot(224)
plt.scatter(set_4[0][:, 0], set_4[0][:, 1], c=set_4[1], cmap='viridis')
plt.title('Zestaw danych 1.4.')

plt.show()
```

```python
model_1 = KMeans(n_clusters=3, random_state=3042019).fit(set_1[0])
model_2 = KMeans(n_clusters=8, random_state=42).fit(set_2[0])
model_3 = KMeans(n_clusters=2, random_state=0).fit(set_3[0])
model_4 = KMeans(n_clusters=2, random_state=0).fit(set_4[0])
```

```python
model_2.cluster_centers_
```

```python
# Dane wej≈õciowe
X_set_1 = set_1[0]

# Etykiety klastr√≥w przypisane przez model
labels_set_1 = model_1.labels_

# Pozycje centroid√≥w klastr√≥w
centers_set_1 = model_1.cluster_centers_

# Tworzenie wykresu
plt.figure(figsize=(8, 6))
plt.scatter(X_set_1[:, 0], X_set_1[:, 1], c=labels_set_1, cmap='viridis')
plt.scatter(centers_set_1[:, 0], centers_set_1[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.title('Model K-means dla P1 (3 klastry)')
plt.legend()
plt.show()
```

```python
# Zbi√≥r danych 1.1
model_1 = KMeans(n_clusters=3, random_state=3042019).fit(set_1[0])

# Zbi√≥r danych 1.2
model_2 = KMeans(n_clusters=8, random_state=42).fit(set_2[0])

# Zbi√≥r danych 1.3
model_3 = KMeans(n_clusters=2).fit(set_3[0])

# Zbi√≥r danych 1.4
model_4 = KMeans(n_clusters=2).fit(set_4[0])

# Tworzenie subplot√≥w
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Wizualizacja modelu 1
axs[0, 0].scatter(set_1[0][:, 0], set_1[0][:, 1], c=model_1.labels_, cmap='viridis')
axs[0, 0].scatter(model_1.cluster_centers_[:, 0], model_1.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')
axs[0, 0].set_title('Model K-means dla P1 (3 klastry)')

# Wizualizacja modelu 2
axs[0, 1].scatter(set_2[0][:, 0], set_2[0][:, 1], c=model_2.labels_, cmap='viridis')
axs[0, 1].scatter(model_2.cluster_centers_[:, 0], model_2.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')
axs[0, 1].set_title('Model K-means dla P2 (8 klastr√≥w)')

# Wizualizacja modelu 3
axs[1, 0].scatter(set_3[0][:, 0], set_3[0][:, 1], c=model_3.labels_, cmap='viridis')
axs[1, 0].scatter(model_3.cluster_centers_[:, 0], model_3.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')
axs[1, 0].set_title('Model K-means dla P3 (2 klastry)')

# Wizualizacja modelu 4
axs[1, 1].scatter(set_4[0][:, 0], set_4[0][:, 1], c=model_4.labels_, cmap='viridis')
axs[1, 1].scatter(model_4.cluster_centers_[:, 0], model_4.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')
axs[1, 1].set_title('Model K-means dla P4 (2 klastry)')

# Wy≈õwietlenie wszystkich subplot√≥w
plt.tight_layout()
plt.show()
```

```python
#4. Zastosowanie algorytmu k-≈õrednich dla liczby grup [2,3,4,5,6,7,10,15] dla P2.

# Dane wej≈õciowe
X_set_2 = set_2[0]

# Lista liczby klastr√≥w, dla kt√≥rych bƒôdziemy stosowaƒá k-≈õrednich
num_clusters_list = [2, 3, 4, 5, 6, 7, 10, 15]

# Inicjalizacja pustej listy do przechowywania wynikowych modeli
kmeans_models = []

# Iteracja po r√≥≈ºnych liczbach klastr√≥w
for num_clusters in num_clusters_list:
    model = KMeans(n_clusters=num_clusters, random_state=42).fit(X_set_2)
    kmeans_models.append(model)

# Wykresy wynikowe
plt.figure(figsize=(12, 8))

for i, model in enumerate(kmeans_models):
    plt.subplot(2, 4, i + 1)
    plt.scatter(X_set_2[:, 0], X_set_2[:, 1], c=model.labels_, cmap='viridis')
    plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')
    plt.title(f'K-means dla P2 ({num_clusters_list[i]} klastr√≥w)')

plt.tight_layout()
plt.show()
```

```python
#5. Dob√≥r optymalnej liczby grup dla P2 (metoda ≈Çokcia).

# Zbi√≥r danych 1.2
set_2 = make_blobs(n_samples=2000, centers=8, cluster_std=2.5, random_state=42)

# Dane wej≈õciowe
X_set_2 = set_2[0]

# Lista r√≥≈ºnych warto≈õci liczby klastr√≥w
num_clusters_range = range(1, 16)  # Pr√≥bujemy od 1 do 15 klastr√≥w

# Inicjalizacja listy do przechowywania wynikowych sumy kwadrat√≥w odleg≈Ço≈õci
inertia_values = []

# Iteracja po r√≥≈ºnych liczbach klastr√≥w
for num_clusters in num_clusters_range:
    model = KMeans(n_clusters=num_clusters, random_state=42).fit(X_set_2)
    inertia_values.append(model.inertia_)

# Wykres "≈Çokcia"
plt.figure(figsize=(8, 6))
plt.plot(num_clusters_range, inertia_values, marker='o')
plt.title('Metoda "≈Åokcia" dla P2')
plt.xlabel('Liczba klastr√≥w')
plt.ylabel('Suma kwadrat√≥w odleg≈Ço≈õci')
plt.grid(True)
plt.show()
```
